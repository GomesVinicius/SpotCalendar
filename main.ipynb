```python
import os
import spotipy
import pandas as pd
from dotenv import load_dotenv
from spotipy.oauth2 import SpotifyOAuth
```

```python
load_dotenv()

CLIENT_ID = os.getenv("CLIENT_ID")
CLIENT_SECRET = os.getenv("CLIENT_SECRET")
```

```python
print(CLIENT_ID)
```

```python
sp = spotipy.Spotify(
    auth_manager=SpotifyOAuth(
        client_id=CLIENT_ID,
        client_secret=CLIENT_SECRET,
        redirect_uri="http://127.0.0.1:8888",
        scope=["playlist-read-private", 'playlist-read-collaborative']
    )
)
```

```python
display_name = sp.current_user()['display_name']
```

```python
user_id = sp.current_user()['id']
```

```python
playlists = sp.current_user_playlists()
```

```python
playlists = sp.current_user_playlists()
playlist_songs = []

for playlist in playlists["items"]:
    # print(playlist["name"], playlist['id'])
    playlist_id = playlist['id']

    if display_name == playlist['owner']['display_name']:
        try:
            items = sp.playlist_items(playlist_id=playlist_id)
            
            for idx, item in enumerate(items['items'], start=1):
                # print(item)
                # time.sleep(50)
                playlist_songs.append({
                    'id': idx,
                    'name':  item['item']['name'],
                    'artists': ', '.join(
                            dict.fromkeys(
                                artist['name']
                                for artist in item['item']['artists']
                            )
                        ),
                    'playlist': playlist['name'],
                    'added_at': item['added_at'],
                    'url_song': item['item']['external_urls']['spotify'],
                    'image': item['item']['album']['images'][1]['url']
                })
        except:
            pass
```

```python
df = pd.DataFrame(playlist_songs)
```

```python
df['added_at'] = pd.to_datetime(df['added_at'])
df['added_at'] = df['added_at'].dt.tz_localize(None)

df_mes_5 = df[df['added_at'].dt.month == 5]
```
